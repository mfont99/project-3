```python
# Dependencies
import pandas as pd
from pathlib import Path
import os
import csv
```

```python
# Save path to the data set as a variable; why does the relative path never work! have to rely on full path! ..............ohh wait it worked -- had to add directory onto path/ so three levels to get to actual file

Url = Path("../project-3/Resources/NCHS_-_Drug_Poisoning_Mortality_by_County__United_States_20240530.csv")
```

```python
#read csv file
data_file = pd.read_csv(Url)
data_file.head()
```

```python
#df the file
Odata_file_df = pd.DataFrame(data_file)
Odata_file_df
```

```python
#rename last column
data_file_df = Odata_file_df.rename(columns={"Estimated Age-adjusted Death Rate, 16 Categories (in ranges)":"Est. Age-Adj death rate (in ranges)"})
data_file_df.head()
```

```python

# export csv file version
data_file_csv = Path("../project-3/Resources/data_file.csv")
data_file_df.to_csv(data_file_csv, index=False)

# export json file version
data_file_json = Path("../project-3/Resources/data_file.json")
data_file_df.to_json(data_file_json, orient="records")
```

```python
#get entries with highest Age adj Death rate counts  
filtered_df = data_file_df[data_file_df['Est. Age-Adj death rate (in ranges)'].str.contains('>30')]
filtered_df.head()
```

```python
#just confirming the total count of entries
filtered_df["Est. Age-Adj death rate (in ranges)"].value_counts()
```

```python
#export filtered to csv

# Define the path to the resources folder
output_file = Path("../project-3/Resources/filtered_data.csv")

filtered_df.to_csv(output_file, index=False)
```

```python
#import geo file for counties' lang and lat
geo_df = pd.read_csv(Path("../project-3/Resources/us_county_latlng.csv"))
geo_df
```

```python
geo_df = geo_df.drop(columns={"name"})
geo_df
```

```python
geo_df = geo_df.rename(columns={"fips_code":"FIPS"})
geo_df.head()
```

```python
merged_df = pd.merge(filtered_df, geo_df, on='FIPS', how='left')
merged_df.head()
```

```python
#how many Nan do we have  ----------- FIPcode for this one AK county not in geo data (wasnt official county by then?) 
nan_count = merged_df['lng'].isna().sum()
nan_count
```

```python
#export merged data inot csv 
output_file2 = Path("../project-3/Resources/merged_data.csv")
merged_df.to_csv(output_file2, index=False)
```

```python
# Save the merged data with latitude and longitude to a JSON file
output_json_file = Path("../project-3/Resources/merged_data.json")
merged_df.to_json(output_json_file, orient='records')
```

```python
import json
# import json file
with open('../project-3/Resources/merged_data.json', 'r') as f:
    data = json.load(f)

# Create a GeoJSON structure
geojson = {
    "type": "FeatureCollection",
    "features": []
}

# Map the JSON data to GeoJSON format
for entry in data:
    if entry['lng'] is not None and entry['lat'] is not None:
        feature = {
            "type": "Feature",
            "geometry": {
                "type": "Point",
                "coordinates": [entry['lng'], entry['lat']]
            },
            "properties": {
                "FIPS": entry["FIPS"],
                "Year": entry["Year"],
                "State": entry["State"],
                "ST": entry["ST"],
                "FIPS State": entry["FIPS State"],
                "County": entry["County"],
                "Population": entry["Population"],
                "Est. Age-Adj death rate (in ranges)": entry["Est. Age-Adj death rate (in ranges)"]
            }
        }
        geojson['features'].append(feature)

# Save the GeoJSON data to a file
with open('../project-3/Resources/merged_data.geojson', 'w') as f:
    json.dump(geojson, f, indent=2)
```
