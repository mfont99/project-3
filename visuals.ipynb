```python
# Import the dependencies.
from pathlib import Path
from sqlalchemy import create_engine
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
link = Path("../project-3/Resources/filtered_data.csv")
filter_file = pd.read_csv(link, dtype={'FIPS': str, 'Year': int, 'State': str, 'ST': str, 'FIPS State': str, 'County': str, 'Population': int, 'Est. Age-Adj death rate (in ranges)': str})
```

```python
county_counts = filter_file.groupby(['State', 'Year']).size().reset_index(name='count')
county_counts
```

```python
# Set up the plot
plt.figure(figsize=(34, 27))

# Create the distribution plot
sns.barplot(data=county_counts, x='State', y='count', hue='Year')

# Add titles and labels
plt.title('Number of Counties with Age-Adjusted Death Rates >30 by State and Year')
plt.xlabel('State')
plt.ylabel('Number of Counties')
plt.xticks(rotation=90)
plt.legend(title='Year')
plt.tight_layout()
# Show the plot
plt.show()
```
